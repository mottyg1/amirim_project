```python
from collections import Counter
import itertools
import matplotlib.pyplot as plt
import igraph as ig
from utils import random_walk_until_return

def kroncker(m,n):
    return 1 if m==n else 0

def simulate(g, samples=100000):
    c = Counter([len(random_walk_until_return(g))-1 for _ in range(samples)])
    return c.keys(), [x/samples for x in c.values()]

def plot_simulation_and_theory(g, theory, title):
    x, y = simulate(g)
    y_theory = [theory(n) for n in x]

    plt.plot(x,y,'ob',label='simulation', markerfacecolor='none')
    plt.plot(x,y_theory,'xr',label='theory')
    plt.title(title)
    plt.yscale('log')
    plt.legend()
```

```python
t3 = ig.Graph()
t3.add_vertices(3)
t3.add_edges([(0,1),(1,2)])
t3_theory = lambda n: 1/3*(2**(1-n/2)+kroncker(2,n))

plot_simulation_and_theory(t3, t3_theory, 't3')
```

```python
t4_1 = ig.Graph()
t4_1.add_vertices(4)
t4_1.add_edges([(0,1),(1,2),(2,3)])
t4_1_theory = lambda n: 2**(-2-n/2)+2**(-n)*3**(-2+n/2)+5/12*kroncker(2,n)

plot_simulation_and_theory(t4_1, t4_1_theory, 't4_1')
```

```python
t4_2 = ig.Graph()
t4_2.add_vertices(4)
t4_2.add_edges([(0,1),(0,2),(0,3)])
t4_2_theory = lambda n: 1/8*(2**(n/2)*3**(1-n/2)+2*kroncker(2,n))

plot_simulation_and_theory(t4_2, t4_2_theory, 't4_2')
```

```python
t5_3 = ig.Graph()
t5_3.add_vertices(5)
t5_3.add_edges([(0,1),(0,2),(0,3),(0,4)])
t5_3_theory = lambda n: 1/5*(2**(2-n)*3**(-1+n/2)+kroncker(2,n))

plot_simulation_and_theory(t5_3, t5_3_theory, 't5_3')
```

```python
t4_1_x, t4_1_y = simulate(t4_1)
t4_2_x, t4_2_y = simulate(t4_2)

plt.plot(t4_1_x,t4_1_y,'xb',label='t4_1', markerfacecolor='none')
plt.plot(t4_2_x,t4_2_y,'or',label='t4_2', markerfacecolor='none')
plt.legend
```

```python
t5_1 = ig.Graph()
t5_1.add_vertices(5)
t5_1.add_edges([(0,1),(1,2),(2,3),(3,4)])

t5_2 = ig.Graph()
t5_2.add_vertices(5)
t5_2.add_edges([(0,1),(0,2),(2,3),(2,4)])
```

```python
t5_1_x, t5_1_y = simulate(t5_1)
t5_2_x, t5_2_y = simulate(t5_2)
t5_3_x, t5_3_y = simulate(t5_3)

plt.plot(t5_1_x,t5_1_y,'xb',label='t5_1', markerfacecolor='none')
plt.plot(t5_2_x,t5_2_y,'or',label='t5_2', markerfacecolor='none')
plt.plot(t5_3_x,t5_3_y,'og',label='t5_3', markerfacecolor='none')
plt.legend()
plt.xlim(0,10)
```
